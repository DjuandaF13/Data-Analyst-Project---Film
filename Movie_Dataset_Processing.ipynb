```python
import pandas as pd
```

```python
df = pd.read_csv('/content/movies_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['ReleaseDate'] = pd.to_datetime(df['ReleaseDate'])
```

```python
df.info()
```

```python
df['RottenTomatoesScore'] = pd.to_numeric(df['RottenTomatoesScore'], errors='coerce')
df['IMDbRating'] = pd.to_numeric(df['IMDbRating'], errors='coerce')
```

```python
df.info()
```

```python
df['ProfitUSD'] = df['Global_BoxOfficeUSD'] - df['BudgetUSD']
```

```python
df.info()
```

```python
df.head()
```

```python
df[['BudgetUSD', 'US_BoxOfficeUSD', 'Global_BoxOfficeUSD', 'Opening_Day_SalesUSD', 'One_Week_SalesUSD',
    'RottenTomatoesScore', 'IMDbRating', 'NumVotesIMDb', 'NumVotesRT']] = df[['BudgetUSD', 'US_BoxOfficeUSD',
    'Global_BoxOfficeUSD', 'Opening_Day_SalesUSD', 'One_Week_SalesUSD', 'RottenTomatoesScore', 'IMDbRating',
    'NumVotesIMDb', 'NumVotesRT']].fillna(0)
```

```python
df['Dekade'] = (df['ReleaseYear'] // 10 * 10).astype(str) + 's'
```

```python
print("\n--- Data Setelah Pembersihan & Transformasi (5 baris pertama) ---")
print(df.head())
```

```python
# 1. Analisis Performa per Genre
print("\n--- Analisis Performa Rata-Rata per Genre ---")
genre_performance = df.groupby('Genre').agg(
    Avg_Profit_USD=('ProfitUSD', 'mean'),
    Avg_IMDb_Rating=('IMDbRating', 'mean'),
    Avg_Budget_USD=('BudgetUSD', 'mean'),
    Total_Movies=('MovieID', 'count')
).sort_values('Avg_Profit_USD', ascending=False)
print(genre_performance)
```

```python
# 2. Analisis Tren Tahunan
print("\n--- Analisis Tren Tahunan ---")
yearly_trends = df.groupby('ReleaseYear').agg(
    Avg_Global_BoxOffice=('Global_BoxOfficeUSD', 'mean'),
    Avg_Budget_USD=('BudgetUSD', 'mean'),
    Avg_IMDb_Rating=('IMDbRating', 'mean')
)
print(yearly_trends.head())
```

```python
# 3. Analisis Korelasi Rating
print("\n--- Korelasi Antara Rating IMDb dan Rotten Tomatoes ---")
correlation_rating = df[['IMDbRating', 'RottenTomatoesScore']].corr()
print(correlation_rating)
```

```python
# 4. Analisis Berdasarkan Dekade
print("\n--- Analisis Berdasarkan Dekade ---")
decade_summary = df.groupby('Dekade').agg(
    Avg_Profit_USD=('ProfitUSD', 'mean'),
    Avg_IMDb_Rating=('IMDbRating', 'mean'),
    Total_Movies=('MovieID', 'count')
)
print(decade_summary)
```

```python
# 5. Analisis Aktor Teratas (Berdasarkan Total Pendapatan)
print("\n--- 10 Aktor Teratas Berdasarkan Total Pendapatan ---")
top_10_actors = df.groupby('LeadActor').agg(
    Total_BoxOffice_USD=('Global_BoxOfficeUSD', 'sum')
).sort_values('Total_BoxOffice_USD', ascending=False).head(10)
print(top_10_actors)
```

```python
# 6. Analisis Sutradara Teratas (Berdasarkan Total Pendapatan)
print("\n--- 10 Sutradara Teratas Berdasarkan Total Pendapatan ---")
top_10_directors = df.groupby('Director').agg(
    Total_BoxOffice_USD=('Global_BoxOfficeUSD', 'sum')
).sort_values('Total_BoxOffice_USD', ascending=False).head(10)
print(top_10_directors)
```

```python
# Simpan DataFrame yang sudah diolah ke file CSV baru
output_file_name = 'cleaned_movie_data_for_looker.csv'
df.to_csv(output_file_name, index=False)

print(f"\n--- Data yang Diolah Telah Disimpan ---")
print(f"File '{output_file_name}' sudah dibuat dan siap diunduh.")
```
